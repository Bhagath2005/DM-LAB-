```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler

data = pd.read_csv('employee_data.csv')

data['Age'] = data['Age'].fillna(data['Age'].median())
data['Salary'] = data['Salary'].fillna(data['Salary'].median())
data['Job_Satisfaction'] = data['Job_Satisfaction'].fillna(data['Job_Satisfaction'].median())

min_max_scaler = MinMaxScaler()
data[['Age', 'Salary']] = min_max_scaler.fit_transform(data[['Age', 'Salary']])

data['Age_manual'] = (data['Age'] - data['Age'].min()) / (data['Age'].max() - data['Age'].min())
data['Salary_manual'] = (data['Salary'] - data['Salary'].min()) / (data['Salary'].max() - data['Salary'].min())

standard_scaler = StandardScaler()
data[['Job_Satisfaction', 'Work_Hours_Per_Week']] = standard_scaler.fit_transform(data[['Job_Satisfaction', 'Work_Hours_Per_Week']])

data['Job_Satisfaction_manual'] = (data['Job_Satisfaction'] - data['Job_Satisfaction'].mean()) / data['Job_Satisfaction'].std()
data['Work_Hours_Per_Week_manual'] = (data['Work_Hours_Per_Week'] - data['Work_Hours_Per_Week'].mean()) / data['Work_Hours_Per_Week'].std()

print(data)
```
